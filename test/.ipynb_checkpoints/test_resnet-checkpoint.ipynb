```python
import h5py
import json
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import model_from_json
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.imagenet_utils import preprocess_input

# ─────────────────────────────────────────────────────────────
#  Compatibility fix: patch 'batch_shape' → 'batch_input_shape'
#  (model was saved with a newer TF; we're running TF 2.4.1)
# ─────────────────────────────────────────────────────────────
def fix_batch_shape(config):
    """Recursively rename batch_shape to batch_input_shape in layer configs."""
    if isinstance(config, dict):
        if 'batch_shape' in config:
            config['batch_input_shape'] = config.pop('batch_shape')
        for v in config.values():
            fix_batch_shape(v)
    elif isinstance(config, list):
        for item in config:
            fix_batch_shape(item)

def load_model_compat(filepath):
    """Load a Keras .h5 model with batch_shape compatibility fix."""
    with h5py.File(filepath, 'r') as f:
        # Read and patch the model config
        raw_config = f.attrs.get('model_config')
        if raw_config is None:
            raise ValueError('No model_config found in file.')
        model_config = json.loads(raw_config.decode('utf-8'))
        fix_batch_shape(model_config)

        # Rebuild the model from patched config
        model = model_from_json(json.dumps(model_config))

        # Load weights
        from tensorflow.python.keras.saving.hdf5_format import load_weights_from_hdf5_group
        load_weights_from_hdf5_group(f['model_weights'], model.layers)

    print('✅ Model loaded successfully!')
    return model

# Load the ResNet model
model = load_model_compat('./model_blood_group_detection_resnet.h5')
```

```python
# ─────────────────────────────────────────────────────────────
#  Run a prediction on a fingerprint image
# ─────────────────────────────────────────────────────────────

# Class labels (alphabetical — same order Keras ImageDataGenerator uses)
labels = {0: 'A+', 1: 'A-', 2: 'AB+', 3: 'AB-', 4: 'B+', 5: 'B-', 6: 'O+', 7: 'O-'}

# ← Change this path to any fingerprint image you want to test
img_path = 'O- blood group.BMP'

# Detect input size expected by the model
input_shape = model.input_shape[1:3]  # e.g. (224, 224) or (256, 256)
print(f'Model expects input size: {input_shape}')

# Load and preprocess
img = image.load_img(img_path, target_size=input_shape)
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# Predict
result = model.predict(x)
predicted_class = np.argmax(result)
predicted_label = labels[predicted_class]
confidence = result[0][predicted_class] * 100

# Display
plt.figure(figsize=(4, 4))
plt.imshow(image.load_img(img_path, target_size=input_shape))
plt.axis('off')
plt.title(f'Predicted: {predicted_label}  |  Confidence: {confidence:.2f}%', fontsize=13)
plt.tight_layout()
plt.show()

print(f'\n🩸 Blood Group: {predicted_label}')
print(f'📊 Confidence : {confidence:.2f}%')
print(f'\nAll class probabilities:')
for i, prob in enumerate(result[0]):
    print(f'  {labels[i]:>4s}: {prob*100:.2f}%')
```
